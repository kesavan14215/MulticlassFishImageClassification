```python

%%writefile app.py
import streamlit as st
import tensorflow as tf
import numpy as np
from PIL import Image
import io

class FishClassifierApp:
    def __init__(self,class_names):       # model_path,
        # Load your saved model and class names
        self.class_names = class_names

    def run(self):
        self.base_models = {
                            'Custom CNN' : '/content/drive/MyDrive/AI ML Practice/MulticlassFishImageClassification/models/CNN_Custom_model.h5',
                            #'Efficientnet' : '/content/drive/MyDrive/AI ML Practice/MulticlassFishImageClassification/models/VGG16_model.h5',
                            #'Inception' : '/content/drive/MyDrive/AI ML Practice/MulticlassFishImageClassification/models/ResNet50_model.h5',
                            #'Mobilenet' : '/content/drive/MyDrive/AI ML Practice/MulticlassFishImageClassification/models/MobileNet_model.h5',
                            #'Resnet50' : '/content/drive/MyDrive/AI ML Practice/MulticlassFishImageClassification/models/InceptionV3_model.h5',
                            #'Vgg16' : '/content/drive/MyDrive/AI ML Practice/MulticlassFishImageClassification/models/EfficientNetB0_model.h5',
                           }

        st.title('Fish Species Classifier')
        st.write('Upload an image of a fish to classify its species')

        uploaded_file = st.file_uploader(
            "Choose an image...",
            type=['jpg', 'jpeg', 'png']
        )


        for key,value in self.base_models.items():
          self.model = tf.keras.models.load_model(value)
          if uploaded_file is not None:
              # Display the uploaded image
              image = Image.open(uploaded_file)
              st.image(image, caption='Uploaded Image', use_container_width=True)

              # Preprocess the image
              img_array = self.preprocess_image(image)

              # Make prediction
              prediction = self.model.predict(img_array)

              # Get top 3 predictions
              top_3_idx = prediction[0].argsort()[-3:][::-1]
              top_3_predictions = prediction[0][top_3_idx]
              top_3_classes = [self.class_names[i] for i in top_3_idx]

              # Display results
              st.subheader('Prediction Results:'+key)
              for class_name, confidence in zip(top_3_classes, top_3_predictions):
                  st.write(f"{class_name}: {confidence*100:.2f}%")
                  st.progress(float(confidence))

    @staticmethod
    def preprocess_image(image, target_size=(224, 224)):
        # Resize image
        image = image.resize(target_size)

        # Convert to array and normalize
        img_array = tf.keras.preprocessing.image.img_to_array(image)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = img_array / 255.0
        return img_array

if __name__ == '__main__':



     # Replace with your actual class names
    CLASS_NAMES = ['animal fish', 'animal fish bass', 'fish sea_food black_sea_sprat', 'fish sea_food gilt_head_bream', 'fish sea_food hourse_mackerel', 'fish sea_food red_mullet', 'fish sea_food red_sea_bream', 'fish sea_food sea_bass', 'fish sea_food shrimp', 'fish sea_food striped_red_mullet', 'fish sea_food trout']

    #app = FishClassifierApp(MODEL_PATH, CLASS_NAMES)
    app = FishClassifierApp( CLASS_NAMES)
    app.run()
```
